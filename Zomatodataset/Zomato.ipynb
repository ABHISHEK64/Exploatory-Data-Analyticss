```python
df_Merger= pd.merge(df, df2, on='Country Code', how='left')
df_Merger.head()
```

```python
###Maximum No resturant in witch country
country_Name=df_Merger['Country'].value_counts().index
values=df_Merger['Country'].value_counts().values
```

```python
sns.barplot(x = country_Name[:5],y=values[:5])
```

##group by country have particular resturant

```python
df_Merger.head()
```

```python
rating_val=df_Merger.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
rating_val
```

```python
plt.rcParams['figure.figsize']=(13,7)
sns.barplot(x='Aggregate rating', y='Rating Count',data=rating_val,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

# Observation
1.Maximum People Rated 0.0 Not Rated
2.If ignore Not Rated then Most people give Avergare or Good Rating
3.least people give poor and excellent rating

```python
### Top 10 Cusine Served by Zomato
df_Merger.groupby('Cuisines').size().sort_values(ascending=False)
```
